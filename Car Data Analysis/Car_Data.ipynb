```python
import pandas as pd
data = pd.read_csv(r"C:\Users\yuvra\Downloads\Car_Data.csv")
data.shape
```

```python
data.isnull().sum()
```

```python
data.head() 
```

```python
data.EngineSize.fillna(data.EngineSize.mean(),inplace=True)
```

```python
data.EngineSize.isnull().sum()
```

```python
data.head(2)
```

```python
data['MSRP'] = data['MSRP'].replace(r'[\$,]', '', regex=True).astype(float)
```

```python
data.head(2)
```

```python
data.MSRP.fillna(data.MSRP.mean(),inplace=True)
```

```python
data.isnull().sum()
```

```python
data['MSRP'] = data['MSRP'].apply(lambda x: f"${x:,.2f}")
```

```python
data.head()
```

```python
data.Horsepower.fillna(data.Horsepower.mean(),inplace=True)
```

```python
data.Cylinders.fillna(data.Cylinders.mean(),inplace=True)
```

```python
data.MPG_City.fillna(data.MPG_City.mean(),inplace=True)
```

```python
data.Weight.fillna(data.Weight.mean(),inplace=True)
```

```python
data.Wheelbase.fillna(data.Wheelbase.mean(),inplace=True)
data.Length.fillna(data.Length.mean(),inplace=True)
data.MPG_Highway.fillna(data.MPG_Highway.mean(),inplace=True)
```

```python
data['Invoice'] = data['Invoice'].replace(r'[\$,]', '', regex=True).astype(float)
data.Invoice.fillna(data.Invoice.mean(),inplace=True)
data['Invoice'] = data['Invoice'].apply(lambda x: f"${x:,.2f}")
```

```python
data['Make'].fillna(data['Make'].mode()[0], inplace=True)
data['Model'].fillna(data['Model'].mode()[0], inplace=True)
data['Type'].fillna(data['Type'].mode()[0], inplace=True)
data['Origin'].fillna(data['Origin'].mode()[0], inplace=True)
data['DriveTrain'].fillna(data['DriveTrain'].mode()[0], inplace=True)
```
